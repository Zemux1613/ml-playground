```python
import pickle
import numpy as np
```

```python
loaded_model = ""
# Load the model from the file using pickle
with open('banana_quality.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
i = 0
# Generate a random NumPy array with 7 elements
while i < 100:
    random_array = np.random.uniform(-4, 8, size=(1, 7))
    prediction = loaded_model.predict(random_array)
    print(f"Predicted class for {random_array} is {'Good' if prediction == 1 else 'Bad'}")
    i = i + 1
```
