```python
import numpy as np
import json
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import sklearn
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
content = open('data.json', encoding="utf-8", mode="r").read()
```

```python
data = json.loads(content)
```

```python
# Sample data (positive examples with correct grammar)
positive_sentences = data["positiv"]

# Sample data (negative examples with incorrect grammar)
negative_sentences = data["negativ"]
```

```python
# Labeling the data
positive_labels = [1] * len(positive_sentences)
negative_labels = [0] * len(negative_sentences)
```

```python
# Combine positive and negative examples
all_sentences = positive_sentences + negative_sentences
all_labels = positive_labels + negative_labels
```

```python
# Tokenize the sentences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(all_sentences)
total_words = len(tokenizer.word_index) + 1

# Convert sentences to sequences
sequences = tokenizer.texts_to_sequences(all_sentences)
padded_sequences = pad_sequences(sequences)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, all_labels, test_size=0.2, random_state=42)
```
