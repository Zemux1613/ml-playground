```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

### Hilfsfunktionen

```python
def plot(X,y):
    # Die Daten werden nach ihren Labeln (0 oder 1) aufgeteilt
    x_1 = X[:, 0]
    x_2 = X[:, 1]
    # Die einzelnen Elemente werden als Punkte dargestellt, wobei die Klasse 0 rot und die Klasse 1 blau dargestellt wird. 
    plt.scatter(X[y==0][:,0], X[y==0][:,1], c="red", edgecolors='black')
    plt.scatter(X[y==1][:,0], X[y==1][:,1], c="blue", edgecolors='black')
    # Nun verstecken wir die Beschriftung der X- und Y- Achse 
    plt.xticks([]) 
    plt.yticks([]) 

def plot_contour(X,y,model, bx, by):
    # Grenzen und Auflösung für das Schaubild festlegen
    x_min, x_max  = bx
    y_min, y_max = by
    res = 0.01
    # Das Modell macht Vorhersagen für viele Punkte im Hintergrund, um die Decisioun Boundary zu zeigen
    grid_x, grid_y = np.meshgrid(np.arange(x_min, x_max, res),
                                 np.arange(y_min, y_max, res))
    grid_input = np.c_[grid_x.reshape(-1), grid_y.reshape(-1)]
    grid_out = model.predict(grid_input).reshape(grid_x.shape)
    plt.contourf(grid_x, grid_y, grid_out, alpha=0.5,cmap='RdBu')
    # Hier werden wieder die Elemente im Datensatz als Punkte dargestellt (siehe Funktion plot())
    x_1 = X[:, 0]
    x_2 = X[:, 1]
    alpha_dot=1
    alpha_back=0.8
    plt.scatter(X[y==0][:,0], X[y==0][:,1], c="red", edgecolors='black', label='class 0', alpha=alpha_dot)
    plt.scatter(X[y==1][:,0], X[y==1][:,1], c="blue", edgecolors='black', label='class 1', alpha=alpha_dot)
    plt.xticks([]) 
    plt.yticks([]) 
```

# Unser Datensatz

```python
# Erstelle einen Datensatz mit 1000 Punkten
X,y = make_blobs(n_samples=1000, centers=[(2,2), (10,10)], n_features=2, random_state=0)
# Hier bestimmen wir ein paar Werte, damit unsere Schaubilder gut aussehen
offset = 1
bx = np.min(X[:,0])-offset, np.max(X[:,0])+offset
by = np.min(X[:,1])-offset, np.max(X[:,1])+offset
# Zum Schluss wird der Datensatz in einem Schaubild dargestellt
plot(X,y)
# Anzeigen des Schaubilds
plt.show()
```

```python
# Teile die Daten in Trainings- und Test-Set auf. Das Verhältnis ist 9 zu 1, wir haben also 900 Trainings- und 100 Testpunkte
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1)
# Stelle die Trainingsdaten grafisch dar
plt.title("Trainingsdaten")
plot(X_train,y_train)
# Anzeigen des Schaubilds
plt.show()
```

```python
# Stelle die Testdaten grafisch dar
plt.title("Testdaten")
plot(X_test,y_test)
# Anzeigen des Schaubilds
plt.show()
```

### Logistische Regression

```python
# Erstelle und trainiere ein logistisches Regressionsmodell
reg = LogisticRegression().fit(X_train, y_train)
# Das Modell macht Vorhersagen für die Testdaten
y_hat = reg.predict(X_test)
# Berechne die Genauigkeit der Vorhersagen
num_correct = np.count_nonzero(y_hat == np.array(y_test))
print("Genauigkeit: ", num_correct/len(y_test))
# Stelle die Decision Boundary dar
plot_contour(X,y,reg, bx, by)
# Anzeigen des Schaubilds
plt.show()
```

### Multi Layer Perceptron

```python
# Erstelle ein MLP Model mit 16 Neuronen
mlp = MLPClassifier(hidden_layer_sizes=(16), max_iter=300, random_state=1)
# Trainiere das Netzwerk mit den Trainingsdaten
mlp.fit(X_train, y_train)
# Bestimme die Genauigkeit mithilfe der Testdaten
y_hat = mlp.predict(X_test)
num_correct = np.count_nonzero(y_hat == np.array(y_test))
print("Genauigkeit: ", num_correct/len(y_test))
# Stelle die Decision Boundary dar
plot_contour(X,y,mlp, bx, by)
# Anzeigen des Schaubilds
plt.show()
```

# Ein neuer Datensatz

```python
# Erstelle einen Datensatz aus 1000 Punkten, die halbmondförmig angeordnet sind
X,y = make_moons(n_samples=1000, shuffle=True, noise=0.05, random_state=None)
# Wieder ein paar Optionen für die grafische Darstellung:
offset = 0.3
bx = np.min(X[:,0])-offset, np.max(X[:,0])+offset
by = np.min(X[:,1])-offset, np.max(X[:,1])+offset
# Auch hier teilen wir die Daten wieder in Trainings- und Testdaten auf
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1)
# Stelle die Trainingsdaten grafisch dar
plt.title("Trainingsdaten")
plot(X_train,y_train)
# Anzeigen des Schaubilds
plt.show()
```

```python
# Stelle die Trainingsdaten grafisch dar
plt.title("Testdaten")
plot(X_test,y_test)
# Anzeigen des Schaubilds
plt.show()
```

```python
reg = LogisticRegression().fit(X_train, y_train)
y_hat = reg.predict(X_test)
num_correct = np.count_nonzero(y_hat == np.array(y_test))
print("Genauigkeit: ", num_correct/len(y_test))
plot_contour(X,y,reg, bx, by)
# Anzeigen des Schaubilds
plt.show()
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(16),random_state=1, max_iter=300)
mlp.fit(X_train, y_train)
y_hat = mlp.predict(X_test)
num_correct = np.count_nonzero(y_hat == np.array(y_test))
print("Genauigkeit: ", num_correct/len(y_test))
plot_contour(X,y,mlp, bx, by)
# Anzeigen des Schaubilds
plt.show()
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(32),random_state=1, max_iter=800)
mlp.fit(X_train, y_train)
y_hat = mlp.predict(X_test)
num_correct = np.count_nonzero(y_hat == np.array(y_test))
print("Genauigkeit: ", num_correct/len(y_test))
plot_contour(X,y,mlp, bx, by)
# Anzeigen des Schaubilds
plt.show()
```
