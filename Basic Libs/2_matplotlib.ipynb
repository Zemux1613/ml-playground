# Unser erstes Schaubild - Ein Liniendiagramm

```python
# Erstelle Daten f체r die x-Achse und die y-Achse.
# Die x-Achse sollte einfach aufsteigende Werte enthalten, wobei hier die Werte
# von 0 bis 10 in 0.1-er Schritten angegeben sind.
# Auf der y-Achse werden die Daten aufgetragen, und wir wollen die
# Sinus-Funktion ("math.sin(x)") zeichnen.
x_werte = np.arange(0, 10, 0.1)
y_werte = np.sin(x_werte)

# Gebe zun채chst die ersten drei Werte der Daten zur Kontrolle aus.
print(x_werte[:3], '\n', y_werte[:3])

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Die Daten von oben sollen in den Zeichenbereich gezeichnet werden.
ax.plot(x_werte, y_werte)

# Zeichne das Schaubild.
plt.show()
```

# Mehrere Funktionen im gleichen Diagramm

```python
import matplotlib.pyplot as plt
import numpy as np

# Nun soll die Cosinus-Funktion in denselben Zeichenbereich dazugezeichnet
# werden, wozu beide Kurven 체berinandergelegt werden.
# Dabei sind die Werte auf der x-Achse f체r beide Kurven dieselben.
# Erstelle zun채chst wieder die Daten f체r die x-Achse und die y-Achse.
x_werte = np.arange(0, 10, 0.1)
y_werte_sinus = np.sin(x_werte)
y_werte_cosinus = np.cos(x_werte)

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Die Daten von oben sollen in den Zeichenbereich gezeichnet werden, wobei zwei
# Kurven in denselben Zeichenbereich gezeichnet werden sollen.
# Mit dem Parameter "label" wird ein Name an die Kurve vergeben.
# Mit dem Parameter "color" wird eine Farbe an die Kurve vergeben. Dabei k철nnen
# Farbw철rter oder RGBA-Codes verwendet werden.
ax.plot(x_werte, y_werte_sinus, label="Sinus", color="red")
ax.plot(x_werte, y_werte_cosinus, label="Cosinus", color="#5522aa44")

# Der Zeichenbereich soll einen Titel und eine Bezeichnung f체r die Achsen
# bekommen.
# In Matplotlib kann aller Text mit LaTeX-Befehlen gestaltet werden. So wird
# z. B. der Text zwischen zwei Dollar-Symbolen als mathematische Formel
# interpretiert, wobei Buchstaben kursiv dargestellt werden.
ax.set_title("Schaubild der Sinus- und Cosinus-Funktion")
ax.set_xlabel("$x$-Achse")
ax.set_ylabel("$y$-Achse")

# Die Namen der Kurven, die zuvor z. B. mit label="Sinus" vergeben wurden,
# sollen auch im Zeichenbereich als Legende angezeigt werden.
ax.legend()

# Zeige auerdem ein Raster an, das unterhalb der Kurven gezeichnet werden soll.
ax.set_axisbelow(True)
ax.grid()

# Zeichne das Schaubild.
plt.show()
```

# Gestaltung der Linien

```python
import matplotlib.pyplot as plt
import numpy as np

# Man kann Kurven auch mit verschiedenem Aussehen zeichnen. Dazu kann jeder
# Kurve ein Format zugewiesen werden, nach dem das Schaubild gezeichnet wird.
# Auf matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html
# und matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# sind viele Formatoptionen beschrieben.
# Erstelle zun채chst wieder Beispieldaten.
x_werte = np.arange(0, 10, 0.5)
y_werte_a = np.sin(x_werte) + 6
y_werte_b = np.sin(x_werte) + 4
y_werte_c = np.sin(x_werte) + 2
y_werte_d = np.sin(x_werte) + 0

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Die Daten von oben sollen in den Zeichenbereich gezeichnet werden, wobei
# verschiedene Formate verwendet werden sollen.
# Mit "*" wird keine Linie gezeichnet, sondern nur Sterne an den Datenpunkten.
# Mit "o-" werden Punkte und eine Linie gezeichnet.
# Mit "--" wird eine gestrichelte Linie gezeichnet, aber keine Punkte.
# Mit "linewidth=5.0" wird die gestrichelte Linie dicker gezeichnet.
# Mit "^" wird keine Linie gezeichnet, sondern nur Dreiecke an den Datenpunkten.
# Mit "ms=15" werden die Dreiecke gr철er gezeichnet.
ax.plot(x_werte, y_werte_a, "*", color="red")
ax.plot(x_werte, y_werte_b, "o-", color="cyan")
ax.plot(x_werte, y_werte_c, "--", linewidth=5.0, color="blue")
ax.plot(x_werte, y_werte_d, "^", ms=15, color="orange")

# Zeichne das Schaubild.
plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Man kann Kurven auch nebeneinander zeichnen. Dazu werden mehrere
# Zeichenbereiche ("axes") in einem Schaubild ("figure") neben- oder
# untereinander in einem Raster angeordnet.
# Dabei m체ssen die x-Werte nicht f체r alle Zeichenbereiche gleich sein.
# Erstelle zun채chst wieder Beispieldaten.
x_gerade = np.arange(-2, 8, 0.1)
y_gerade = x_gerade * 1.2 + 5
x_sinus = np.arange(0, 10, 0.1)
y_sinus = np.sin(x_sinus)
x_parabel = np.arange(-5, 5, 0.1)
y_parabel = x_parabel ** 2
x_cosinus = np.arange(-4, 6, 0.1)
y_cosinus = np.cos(x_cosinus)
x_logarithmus = np.arange(0.001, 10, 0.1)
y_logarithmus = np.log(x_logarithmus)
x_exponentialfunktion = np.arange(0, 10, 0.1)
y_exponentialfunktion = np.exp(x_exponentialfunktion)

# Schreibe die Daten in eine zweidimensionale Liste, genau so, wie sie nachher
# angezeigt werden sollen.
x_werte = [[x_gerade, x_sinus, x_parabel],
           [x_cosinus, x_logarithmus, x_exponentialfunktion]]
y_werte = [[y_gerade, y_sinus, y_parabel],
           [y_cosinus, y_logarithmus, y_exponentialfunktion]]
bezeichnungen = [["Gerade", "Sinus", "Parabel"],
                 ["Cosinus", "Logarithmus", "Exponentialfunktion"]]
farben = [["magenta", "green", "blue"],
          ["gray", "cyan", "orange"]]

# Erstelle sechs Zeichenbereiche "ax" ("axes") im Schaubild, so dass in jeder
# Reihe drei davon sind.
anzahl_spalten = 3
anzahl_reihen = 2
fig, ax = plt.subplots(ncols=anzahl_spalten, nrows=anzahl_reihen)

# Die oben generierten Daten sollen nun in je ein eigenes Schaubild gezeichnet
# werden. Dazu wird mit "ax[reihe, spalte]" die richtige Zeichenfl채che
# ausgew채hlt und darin die Kurve gezeichnet.
for reihe in range(anzahl_reihen):
  for spalte in range(anzahl_spalten):
    # Hole die entsprechenden Werte, die Bezeichung und die Farbe der Kurve
    # aus den zuvor erstellten Listen. Jede Kurve in ihrer eigenen
    # Zeichenfl채che kann also eine eigene Bezeichung und Farbe haben.
    x = x_werte[reihe][spalte]
    y = y_werte[reihe][spalte]
    bezeichnung = bezeichnungen[reihe][spalte]
    farbe = farben[reihe][spalte]
    ax[reihe, spalte].plot(x, y, label=bezeichnung, color=farbe)

# Die Namen der Kurven, die zuvor z. B. mit label="Gerade" vergeben wurden,
# sollen auch im Zeichenbereich als Legende angezeigt werden.
fig.legend(bbox_to_anchor=(1.37, 0.34), loc="lower right")

# Lasse ein enges Layout berechnen ("tight"), so dass sich keine Beschriftungen
# 체berlappen.
fig.tight_layout()

# Zeichne das Schaubild.
plt.show()
```

# Anzeigen von Bildern

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import urllib.request
import io

# Mit Matplotlib kann man auch Bilder anzeigen. Das Paket "matplotlib.image"
# stellt die zum Laden von Bildern n철tige Funktionalit채t bereit.
# Importiere zun채chst das Paket und weise ihm das K체rzel "mpimg" zu.
import matplotlib.image as mpimg

# lese ein Beispielbild von einer Online-Quelle und schreibe es als rohe Werte
# in die Variable "bild".
url = "https://www.bw-ki.de/fileadmin/templates/Public/Images/logo_bwki_map.png"
# ffne die URL mit Pillow und konvertiere das Ergebnis in ein NumPy-Array
response = urllib.request.urlopen(url)
img = Image.open(io.BytesIO(response.read()))
bild = mpimg.pil_to_array(img)

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Schalte die Achsen des Schaubilds aus, da man sie bei einem Bild nicht
# ben철tigt.
ax.set_axis_off()

# Zeige das Bild an.
ax.imshow(bild)
plt.show()
```

# Streudiagramme

```python
import matplotlib.pyplot as plt
import numpy as np

# Mit Matplotlib kann man auch Streudiagramme ("Scatterplots") zeichnen. Dabei
# haben die Werte auf der x-Achse, als auch auf der y-Achse eine Bedeutung.
# Weitere Werte-Dimensionen kann man 체ber die Gr철e der Markierungen und deren
# Farbe anzeigen, wobei das dann un체bersichtlich werden kann.
# Als Beispiel sind einige Daten aus L채ndern gegeben: der name des Lands, auf
# welchem Kontinent es liegt, die Bev철lkerungsanzahl, die Fl채che in km^2 und das
# Durchschnittsalter.
# Quelle: https://www.worldometers.info/world-population/population-by-country/
d = """L채ndername,     Kontinent,  Bev철lkerung,  Fl채che,   Durchschnittsalter\n
       China,          Asien,      1439323776,   9388211,  38\n
       Indien,         Asien,      1380004385,   2973190,  28\n
       USA,            Amerika,    331002651,    9147420,  38\n
       Indonesien,     Asien,      273523615,    1811570,  30\n
       Pakistan,       Asien,      220892340,    770880,   23\n
       Brasilien,      Amerika,    212559417,    8358140,  33\n
       Nigeria,        Afrika,     206139589,    910770,   18\n
       Bangladesch,    Asien,      164689383,    130170,   28\n
       Russland,       Asien,      145934462,    16376870, 40\n
       Mexiko,         Amerika,    128932753,    1943950,  29\n
       Japan,          Asien,      126476461,    364555,   48\n
       thiopien,      Afrika,     114963588,    1000000,  19\n
       Philippinen,    Afrika,     109581078,    298170,   26\n
       gypten,        Afrika,     102334404,    995450,   25\n
       Vietnam,        Asien,      97338579,     310070,   32\n
       Kongo,          Afrika,     89561403,     2267050,  17\n
       T체rkei,         Asien,      84339067,     769630,   32\n
       Iran,           Asien,      83992949,     1628550,  32\n
       Deutschland,    Europa,     83783942,     348560,   46\n
       Thailand,       Asien,      69799978,     510890,   40\n
       Grobritannien, Europa,     67886011,     241930,   40\n
       Frankreich,     Europa,     65273511,     547557,   42\n
       Italien,        Europa,     60461826,     294140,   47\n
       Tanzania,       Afrika,     59734218,     885800,   18\n
       S체dafrika,      Afrika,     59308690,     1213090,  28\n"""

# Eine Funktion, in der den Kontinenten eine Farbe zugewiesen wird.
def kontinent_zu_farbe(kontinent):
  undurchsichtigkeit = "a0"
  farben = {"Afrika": "#00ffff" + undurchsichtigkeit,
            "Asien": "#0000ff" + undurchsichtigkeit,
            "Amerika": "#ffa500" + undurchsichtigkeit,
            "Europa": "#ff0000" + undurchsichtigkeit}
  return farben[kontinent]

# Eine Funktion zum Laden der Daten aus der Zeichenkette oben.
def lade_daten(daten_zeichenkette):
  daten = daten_zeichenkette.replace(" ", "").split("\n")
  daten = list(filter(lambda d: d != "", daten))
  daten = [tuple(d.split(",")) for d in daten]
  spalten = np.array(daten[0])
  daten = np.asarray(daten[1:], dtype='object, object, i4, i4, i4')
  return spalten, daten

# Lade die Daten
spalten, daten = lade_daten(d)

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Bereite die die Daten f체r ein Streudiagramm ("Scatterplot") vor. Die
# Landesfl채che soll auf der x-Achse und die Bev철lkerungsanzahl auf der y-Achse
# angezeigt werden. Das Durchschnittsalter der Menschen in den L채ndern wird
# durch die Gr철e des Kreises (Parameter "s") angezeigt, allerdings muss das
# zuvor mit der Wurzelfunktion normiert werden, damit die Kreisfl채che
# proportional zur Bev철lkerungsanzahl ist. Die Farbe des Kreises wird durch den
# Kontinent bestimmt, auf dem das Land liegt.
flaechen = [d[3] for d in daten]
bevoelkerung = np.sqrt([d[2] for d in daten]) / 100
durchschnittsalter = [d[4] for d in daten]
kontinente = [kontinent_zu_farbe(d[1]) for d in daten]

# Das Streudiagramm soll nun die vorbereiteten Daten beinhalten.
# Ein Streudiagramm, bei dem die Gr철en der Kreise eine Bedeutung haben, nennt
# man auch Balldiagramm ("ball chart").
ax.scatter(durchschnittsalter, flaechen, s=bevoelkerung, c=kontinente)
ax.set_xlabel("Durchschnittsalter")
ax.set_ylabel("Landesfl채che")

# Zeichne das Schaubild.
plt.show()
```

# Balkendiagramme

```python
import matplotlib.pyplot as plt
import numpy as np

# Mit Matplotlib kann man auch Balkendiagramme zeichnen. Das macht man, sobald
# kategorische Werte auf der x-Achse aufgetragen sind. Beispielsweise kann man
# die fiktiven Verkaufszahlen von K채sesorten einzeichnen.
namen = ["Weichk채se", "Bergk채se", "Kr채uterk채se", "Raclettek채se"]
werte = [50, 76, 42, 67]

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Die Verkaufszahlen der K채sesorten sollen in ein Balkendiagramm ("bar chart")
# eingezeichnet werden.
ax.bar(namen, werte)
ax.set_title("Verkaufszahlen von K채setypen")
ax.set_xlabel("K채setyp")
ax.set_ylabel("Verkaufszahl")

# Die Bezeichnungen an der x-Achse sollen um 35째 rotiert angezeigt werden.
ax.set_xticklabels(namen, rotation=35, ha='right', rotation_mode='anchor')

# Zeichne das Schaubild.
plt.show()
```

# Text und Markierungen

```python
import matplotlib.pyplot as plt
import numpy as np

# Mit Matplotlib kann man auch Linien und Texte in Schaubilder zeichnen, z. B.
# um Besonderheiten in den Daten zu markieren und zu erkl채ren.

# Definiere Beispielwerte, die in einem Schaubild angezeigt werden sollen.
x_werte = np.arange(0, 10, 0.1)
y_werte = np.sin(x_werte)

# Erstelle ein Schaubild "fig" ("figure") mit dem Zeichenbereich "ax" ("axes").
fig, ax = plt.subplots()

# Die Daten von oben sollen in den Zeichenbereich gezeichnet werden.
ax.plot(x_werte, y_werte)

# Eine orange Linie von Punkt p1 zu Punkt p2 soll gezeichnet werden, wobei die
# Linie ein Pixel stark sein und mit kleinen Strichen abgeschlossen werden soll.
p1 = [0, -0.2]
p2 = [3.1415, -0.2]
ax.plot([p2[0], p1[0]], [p2[1], p1[1]], marker="|", color="orange", lw=1)

# Ein roter Text soll an Position p3 angezeigt werden.
p3 = [0.5, -0.3]
ax.text(p3[0], p3[1], "Halbe Periode", color="r")

# Eine gr체ne horizontale Linie soll von x=0 bis x=6.283 auf H철he y=0.1
# angezeigt werden.
ax.hlines(0.1, 0, 6.283, color="g")

# Ein schwarzer Text soll an Position p4 angezeigt werden.
p4 = [3.5, 0.15]
ax.text(p4[0], p4[1], "Ganze Periode", color="k")

# Eine gelbe vertikale Linie soll von y=-1 bis y=1 an der Stelle x=7.85375
# angezeigt werden.
ax.vlines(7.85375, -1, 1, color="y")

# Zeichne das Schaubild.
plt.show()
```
